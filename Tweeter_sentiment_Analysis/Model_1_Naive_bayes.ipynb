```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
#nltk.download('stopwords')
#nltk.download()
```

```python
train=pd.read_csv('train_E6oV3lV.csv')
```

```python
train.head()
```

```python
train.isna().sum()
```

```python
train.drop('id',axis=1,inplace=True)
```

```python
train.head()
```

```python
train['label'].value_counts()
```

```python
from nltk.stem import WordNetLemmatizer
wnl=WordNetLemmatizer()
```

```python
xtrain = train['tweet']
```

```python
ytrain=train['label']
```

```python
xcorpus=[]
```

```python
for i in range(0, len(xtrain)):
    review = re.sub('[^a-zA-Z]',' ', xtrain[i])  # replaces all the special charater
    review = review.lower()                      # converting everything to normal
    review = nltk.word_tokenize(review)                      # splitig
    
    review = [wnl.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    xcorpus.append(review)
```

```python
# bag of words
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=15000)#max_features=10000
X=cv.fit(xcorpus)
```
